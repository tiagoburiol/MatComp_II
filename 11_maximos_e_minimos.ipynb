## Funções de múltiplas variáveis

Uma função de múltiplas variáveis é uma relação matemática que associa a cada elemento de um subconjunto de $R^n$ um único elemento de $R^m$, $m \geq 1$. 

Essas funções ocorrem com frequência em situações práticas. Por exemplo, o volume de um cilindro circular depende do raio da base e da altura. O custo de um produto depende do custo de cada material utilizado em sua produção, além do custo da fabricação e outros. A temperatura na superfície da terra depende da posição geográfica, data, hora, altitude, e outras variáveis. Uma reação química depende das quantidades dos elementos, temperatura e pressão. A velocidade e direção do vento depende do dia, hora e local, condições climáticas, etc. 

Em ciências e engenharia é comum tratarmos com funções de múltiplas variáveis em um domínio bidimensional ou tridimensional, podendo variar com o tempo ou não. É comum nos referirmos às funções de múltiplas variáveis, especialmente àquelas cujas variáveis dependem do tempo, como funções vetoriais. Vamos tomar alguns casos como exemplos.

Para rodar os códigos dos exemplos a seguir, precisamos importar as bibliotecas `numpy`e `matplotlib`.Podemos fazer isso com as seguintes linhas de código abaixo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Função de duas variáveis reais a uma variável real

É uma relação $f:R^2 \rightarrow R$ que associa cada elemento $(x,y)$
do domínio $D \subset R^2$ a um único elemento de $z \in R$ Podemos escrever 

$$z = f(x,y)$$

assim, os elementos pertencentes a essa função são ponto $(x,y,z)$ do $R^3$. A representação gráfica dessa função pode ser tridimensional, pontos no espaço, ou bidimensional usando, por exemplo, curvas de nível ou mapa de cores. O conjunto imagem pode ser associado a um campo escalar.

```python
f1 = lambda x,y: x**2+y**2
f2 = lambda x,y: x**2-y**2
f3 = lambda x,y: x**4+y**4-4*x*y+1
```

```python
xi, yi = np.meshgrid(np.linspace(-5,5,201),
                     np.linspace(-5,5,201))

#criamos a figura e o subplot 3d
fig = plt.figure(figsize=(13,4))

ax = fig.add_subplot(131, projection='3d')
ax.plot_surface(xi, yi, f1(xi,yi))

ax = fig.add_subplot(132, projection='3d')
ax.plot_surface(xi, yi, f2(xi,yi))

ax = fig.add_subplot(133, projection='3d')
ax.plot_surface(xi, yi, f3(xi,yi))

plt.show()
```

```python
#criamos a figura e o subplot 3d
fig = plt.figure(figsize=(13,4))

ax = fig.add_subplot(131)
c = ax.contour(xi, yi, f1(xi,yi),10)
ax.clabel(c, inline=2, fontsize=8)

ax = fig.add_subplot(132)
c = ax.contour(xi, yi, f2(xi,yi),10)
ax.clabel(c, inline=2, fontsize=8)

ax = fig.add_subplot(133)
niveis = [-0.9, 0, 1, 2,10,50,100, 200, 400, 800]
c = ax.contour(xi, yi, f3(xi,yi),niveis)
ax.clabel(c, inline=2, fontsize=8)

plt.show()
```

Campo gradiente

```python
xi, yi = np.meshgrid(np.linspace(-5,5,21),
                     np.linspace(-5,5,21))

#criamos a figura e o subplot 3d
fig = plt.figure(figsize=(13,4))

ax = fig.add_subplot(131)
u = np.gradient(f1(xi,yi))[1]
v = np.gradient(f1(xi,yi))[0]
plt.quiver(xi,yi,u,v, color='b')
plt.contour(xi, yi, f1(xi,yi),10, colors="r", linewidths=0.4)


ax = fig.add_subplot(132)
u = np.gradient(f2(xi,yi))[1]
v = np.gradient(f2(xi,yi))[0]
plt.quiver(xi,yi,u,v, color='b')
plt.contour(xi, yi, f2(xi,yi),10, colors="r", linewidths=0.4)


ax = fig.add_subplot(133)
u = np.gradient(f3(xi,yi))[1]
v = np.gradient(f3(xi,yi))[0]
plt.quiver(xi,yi,u,v, color='b')
xi, yi = np.meshgrid(np.linspace(-5,5), np.linspace(-5,5))
plt.contour(xi, yi, f3(xi,yi),niveis, colors="r", linewidths=0.4)

plt.show()
```

### Máximos e mínimos de funções de duas variáveis

Uma função de duas variáveis tem um **máximo local** em $(a, b)$ se $f(x,y) \leq f(a,b)$ quando $(x,y)$ está próximo de $(a,b)$. Isso significa que $f(x,y) \leq f(a,b)$ para todos os pontos $(x, y)$ em alguma bola aberta com centro $(a, b)$. O número $f(a, b)$ é chamado **valor máximo local**. Se $f (x, y) \geq f(a, b)$ quando $(x, y)$ está próximo $(a, b)$, então $f$ tem um **mínimo local** em $(a, b)$ e $f (a, b)$ é um **valor mínimo local**.


Se $f$ tem um máximo ou mínimo local em $(a, b)$ e as derivadas parciais de primeira ordem de $f$ existem nesses pontos, então $f_x(a, b)= 0$ e $f_y(a, b)= 0$.

Se as derivadas parciais de $f$ são contínuas nas proximidades de $(a,b)$ e nulas em $(a,b)$ e seja 

$$ D = \left| \begin{array}{cc}
f_{xx} & f_{xy}\\ 
f_{yx} & f_{yy}
\end{array} \right|$$

- Se $D>0$ e $f_{xx}>0$ em $(a,b)$, então $(a,b)$ é um mínimo local.
- Se $D>0$ e $f_{xx}<0$ em $(a,b)$, então $(a,b)$ é um máximo local.
- Se $D<0$ em $(a,b)$, então $(a,b)$ é ponto de sela (ném máximo nem mínimo).
- Se $D=0$ nada se pode concluir.

**Exemplo 1:** Seja $f(x, y)=x^4+y^4-4xy+1$, as suas derivadas parciais são $f_x(x,y)=4x^3-4y$ e $f_y(x,y) = 4y^3-4x$. Igualando essas derivadas parciais a zero, obtemos

$$x^3-y=0 \,\,\,\,\text{e}\,\,\,\, y^3-x=0$$

Substituindo $y=x^3$ da primeira equação na segunda, obtemos $x^9-x=x(x^8-1)=0$, ou seja, as raízes são $x=0$, $1$ ou $-1$. Assim, os três pontos críticos são $(0,0)$, $(1,1)$ e $(-1,-1)$.

As segundas derivadas parciais são $f_{xx}=12x^2$, $f_{xy}=f_{yx}=-4$ e $f_{yy}=12y^2$. Então, $D(x,y)=f_{xx}f_{yy}-(f_{xy})^2 = 144x^2y^2-16$.

Como $D(0,0)=-16$, pelo teste da derivada segunda, o ponto $(0,0)$ é um ponto de sela. Como $D(1,1)=128>0$ e $f_{xx}(1,1)=12>0$, então, $f(1,1)=-1$ é um ponto de mínimo local. Da mesma forma, temos que $D(-1,-1)=128>0$ e $f_{xx}(-1,-1)=12>0$ portanto $f(-1,-1)=-1$ também é um mínimo local.  




A seguir usamos a biblioteca `sympy`para obter as derivadas, calcular os pontos críticos e testar aplicar o teste da derivada segunda.

```python
import sympy as sp
sp.init_printing()

#Define os simbolos e funções
x,y = sp.symbols('x, y', real=True)
z = sp.Function('z')
z = x**4+y**4-4*x*y+1

# Deriva
dzdx = z.diff(x)
dzdy = z.diff(y)

#cria as equações
eq1 = sp.Eq(dzdx,0)
eq2 = sp.Eq(dzdy,0)
pts = sp.solve([eq1,eq2])
pts
```

```python
D = z.diff(x,x)*z.diff(y,y)-z.diff(x,y)**2
```

```python
#substituindo x=-1 e y = -1 na derivada fxx e em D
print('Testando o ponto (0,0):')
print('fxx =',z.diff(x,x).subs([(x, 0), (y, 0)]))
print('D =',D.subs([(x, 0), (y, 0)]))
```

```python
#substituindo x=-1 e y = -1 na derivada fxx e em D
print('Testando o ponto (-1,-1):')
print('fxx =',z.diff(x,x).subs([(x, -1), (y, -1)]))
print('D =',D.subs([(x, -1), (y, -1)]))
```

```python
#substituindo x=1 e y = 1 na derivada fxx e em D
print('Testando o ponto (1,1):')
print('fxx =',z.diff(x,x).subs([(x, 1), (y, 1)]))
print('D =',D.subs([(x, 1), (y, 1)]))
    
```

Gerando uma visualização com curvas de nível nos pontos críticos sobr a superfície.

```python
z= lambda x,y: x**4+y**4-4*x*y+1
```

```python
%matplotlib widget
#então definimos uma malha de pontos (xi,yi)
xi, yi = np.meshgrid(np.linspace(-3, 3, 201), 
                     np.linspace(-3, 3, 201))
zi = z(xi,yi)
zi[zi > 10] = np.nan

#criamos a figura e o subplot 3d
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

#então, plotamos a figura
ax.plot_surface(xi, yi, zi, 
                edgecolor='royalblue', lw=0.2, 
                rstride=4, cstride=4,
                alpha=0.2)

ax.contour(xi, yi, zi, [-1], zdir='x',
                linewidths=0.8, colors='b')
ax.contour(xi, yi, zi, [-1], zdir='y',
                linewidths=0.8, colors='b')

ax.contour(xi, yi, zi, [1], zdir='x',
                linewidths=0.8, colors='r')
ax.contour(xi, yi, zi, [1], zdir='y',
                linewidths=0.8, colors='r')

ax.contour(xi, yi, zi, [-0.9,0,1,3,5,10], zdir='z',
                linewidths=0.8, colors='r', offset=-5)



ax.set_zlim3d(-5, 10)
ax.view_init(20, 120)
plt.show()
```
