#### Função de duas variáveis reais a uma variável real

É uma relação $f:R^2 \rightarrow R$ que associa cada elemento $(x,y)$
do domínio $D \subset R^2$ a um único elemento de $z \in R$ Podemos escrever 

$$z = f(x,y)$$

assim, os elementos pertencentes a essa função são ponto $(x,y,z)$ do $R^3$. A representação gráfica dessa função pode ser tridimensional, pontos no espaço, ou bidimensional usando, por exemplo, curvas de nível ou mapa de cores. O conjunto imagem pode ser associado a um campo escalar.

#### Limites


**Exemplo 1:** Vamos comparar o comportamento das funções

$$
f(x, y)=\frac{\operatorname{sen}\left(x^2+y^2\right)}{x^2+y^2} \quad \text { e } \quad g(x, y)=\frac{x^2-y^2}{x^2+y^2}
$$

quando $x$ e $y$ se aproximam de $0$ ou seja, o ponto $(x, y)$ se aproxima da origem.

```python
xi, yi = np.meshgrid(np.linspace(-5,5,12),
                     np.linspace(-5,5,12))
```

```python
f = lambda x,y: np.sin(x**2+y**2)/(x**2+y**2)
g = lambda x,y: (x**2-y**2)/(x**2+y**2)
```

```python
#print(np.round(xi,2))
#print(np.round(xi,2))
print(np.round(f(xi,yi),2))
```

```python
print(np.round(g(xi,yi),2))
```

Parece que, quando $(x, y)$ se aproxima de $(0,0)$, os valores de $f(x, y)$ se aproximam de 1 , ao passo que os valores de $g(x, y)$ não se aproximam de valor algum. Essa nossa observação baseada em evidências numéricas está correta, e podemos escrever

$$
\lim _{(x, y) \rightarrow(0,0)} \frac{\operatorname{sen}\left(x^2+y^2\right)}{x^2+y^2}=1 \quad \text { e } \quad \lim _{(x, y) \rightarrow(0,0)} \frac{x^2-y^2}{x^2+y^2} \quad \text { não existe }
$$


Para mostrar que a função $g$ não tem limite quando $(x, y)$ tende a $(0,0)$, primeiramente observamos os valores da função se aproximando de $(0,0)$ ao longo do eixo $x$. Nesse caminho, $y=0$ para todo ponto $(x, y)$, de modo que a função se torna $f(x, 0)=x^2 / x^2=1$ para todo $x \neq 0 \mathrm{e}$, portanto,

$$
f(x, y) \rightarrow 1 \quad \text { quando } \quad(x, y) \rightarrow(0,0) \text { ao longo do eixo } x
$$


Agora vamos nos aproximar ao longo do eixo $y$, colocando $x=0$. Então, $f(0, y)=\frac{-y^2}{y^2}=-1$
para todo $y \neq 0$, portanto, para todo $y \neq 0$, portanto,

$$
f(x, y) \rightarrow-1 \quad \text { quando } \quad(x, y) \rightarrow(0,0) \text { ao longo do eixo } y
$$

(veja a Figura 4). Como $f$ tem dois limites diferentes quando $(x, y)$ tende a $(0,0)$ ao longo de duas retas diferentes, o limite não existe. (Isso confirma a conjectura que fizemos

```python
#%matplotlib tk
%matplotlib inline
```

```python
xi, yi = np.meshgrid(np.linspace(-5,5,1000),
                     np.linspace(-5,5,1000))

#criamos a figura e o subplot 3d
fig = plt.figure(figsize=(10,4))

ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(xi, yi, f(xi,yi))

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(xi, yi, g(xi,yi))
plt.show()
```

**Exemplo 2:** ([Fonte](https://integrada.minhabiblioteca.com.br/reader/books/9786555584103/pageid/359)) Se 

$$f(x, y)=\frac{x y}{x^2+y^2}$$, 

será que $\lim _{(x, y) \rightarrow(0,0)} f(x, y) \quad$ existe?

*SOLUÇÃO:* Se $y=0$, então $f(x, 0)=0 / x^2=0$. Portanto, $f(x, y) \rightarrow 0 \quad$ quando $\quad(x, y) \rightarrow(0,0)$ ao longo do eixo $x$. Se $x=0$, então $f(0, y)=0 / y^2=0$, portanto, $f(x, y) \rightarrow 0 \quad$ quando $\quad(x, y) \rightarrow(0,0)$ ao longo do eixo $y$.

Apesar de termos encontrado valores idênticos ao longo dos eixos, não podemos afirmar que o limite exista e seja 0 . Vamos agora nos aproximar de $(0,0)$ ao longo de outra reta; por exemplo, $y=x$. Para todo $x \neq 0$,$f(x, x)=\frac{x^2}{x^2+x^2}=\frac{1}{2}$. Portanto, $\quad f(x, y) \rightarrow \frac{1}{2} \quad$ quando $\quad(x, y) \rightarrow(0,0)$ ao longo de $y=x$.

Como obtivemos valores diferentes para o limite ao longo de cami-nhos diferentes, podemos afirmar que o limite dado não existe.

#### Derivadas parciais

Se $f$ é uma função de duas variáveis, suas derivadas parciais são as funções $f_x$ e $f_y$ definidas por

$$
\begin{aligned}
& f_x(x, y)=\lim _{h \rightarrow 0} \frac{f(x+h, y)-f(x, y)}{h} \\
& f_y(x, y)=\lim _{h \rightarrow 0} \frac{f(x, y+h)-f(x, y)}{h}
\end{aligned}
$$


**Exemplo 3:** ([Fonte](https://integrada.minhabiblioteca.com.br/reader/books/9786555584103/pageid/369)) Se $f(x, y)=4-x^2-2 y^2$, determine $f_x(1,1)$ e $f_y(1,1)$ e interprete esses números como inclinações.

*SOLUÇÃO:* Temos

$$
\begin{array}{ll}
f_x(x, y)=-2 x & f_y(x, y)=-4 y \\
f_x(1,1)=-2 & f_y(1,1)=-4
\end{array}
$$

O gráfico de $f$ é o paraboloide $z=4-x^2-2 y^2$, e o plano vertical $y=1$ intercepta-o na parábola $z=2-x^2, y=1$. A inclinação da reta tangente a essa parábola no ponto $(1,1,1)$ é $f_x(1,1)=-2$. 

Da mesma forma, a curva $C_2$ na qual o plano $x=1$ intercepta o paraboloide é a parábola $z=3-2 y^2, x=1$, e a inclinação da reta tangente em $(1,1,1)$ é $f_y(1,1)=-4$.

### Máximos e mínimos de funções de duas variáveis

Uma função de duas variáveis tem um **máximo local** em $(a, b)$ se $f(x,y) \leq f(a,b)$ quando $(x,y)$ está próximo de $(a,b)$. Isso significa que $f(x,y) \leq f(a,b)$ para todos os pontos $(x, y)$ em alguma bola aberta com centro $(a, b)$. O número $f(a, b)$ é chamado **valor máximo local**. Se $f (x, y) \geq f(a, b)$ quando $(x, y)$ está próximo $(a, b)$, então $f$ tem um **mínimo local** em $(a, b)$ e $f (a, b)$ é um **valor mínimo local**.


Se $f$ tem um máximo ou mínimo local em $(a, b)$ e as derivadas parciais de primeira ordem de $f$ existem nesses pontos, então $f_x(a, b)= 0$ e $f_y(a, b)= 0$.

Se as derivadas parciais de $f$ são contínuas nas proximidades de $(a,b)$ e nulas em $(a,b)$ e seja 

$$ D = \left| \begin{array}{cc}
f_{xx} & f_{xy}\\ 
f_{yx} & f_{yy}
\end{array} \right|$$

- Se $D>0$ e $f_{xx}>0$ em $(a,b)$, então $(a,b)$ é um mínimo local.
- Se $D>0$ e $f_{xx}<0$ em $(a,b)$, então $(a,b)$ é um máximo local.
- Se $D<0$ em $(a,b)$, então $(a,b)$ é ponto de sela (ném máximo nem mínimo).
- Se $D=0$ nada se pode concluir.

**Exemplo 5:** Seja $f(x, y)=x^4+y^4-4xy+1$, as suas derivadas parciais são $f_x(x,y)=4x^3-4y$ e $f_y(x,y) = 4y^3-4x$. Igualando essas derivadas parciais a zero, obtemos

$$x^3-y=0 \,\,\,\,\text{e}\,\,\,\, y^3-x=0$$

Substituindo $y=x^3$ da primeira equação na segunda, obtemos $x^9-x=x(x^8-1)=0$, ou seja, as raízes são $x=0$, $1$ ou $-1$. Assim, os três pontos críticos são $(0,0)$, $(1,1)$ e $(-1,-1)$.

As segundas derivadas parciais são $f_{xx}=12x^2$, $f_{xy}=f_{yx}=-4$ e $f_{yy}=12y^2$. Então, $D(x,y)=f_{xx}f_{yy}-(f_{xy})^2 = 144x^2y^2-16$.

Como $D(0,0)=-16$, pelo teste da derivada segunda, o ponto $(0,0)$ é um ponto de sela. Como $D(1,1)=128>0$ e $f_{xx}(1,1)=12>0$, então, $f(1,1)=-1$ é um ponto de mínimo local. Da mesma forma, temos que $D(-1,-1)=128>0$ e $f_{xx}(-1,-1)=12>0$ portanto $f(-1,-1)=-1$ também é um mínimo local.  




A seguir usamos a biblioteca `sympy`para obter as derivadas, calcular os pontos críticos e testar aplicar o teste da derivada segunda.

```python
import sympy as sp
sp.init_printing()

#Define os simbolos e funções
x,y = sp.symbols('x, y', real=True)
z = sp.Function('z')
z = x**4+y**4-4*x*y+1

# Deriva
dzdx = z.diff(x)
dzdy = z.diff(y)

#cria as equações
eq1 = sp.Eq(dzdx,0)
eq2 = sp.Eq(dzdy,0)
pts = sp.solve([eq1,eq2])
pts
```

```python
D = z.diff(x,x)*z.diff(y,y)-z.diff(x,y)**2
```

```python
#substituindo x=-1 e y = -1 na derivada fxx e em D
print('Testando o ponto (0,0):')
print('fxx =',z.diff(x,x).subs([(x, 0), (y, 0)]))
print('D =',D.subs([(x, 0), (y, 0)]))
```

```python
#substituindo x=-1 e y = -1 na derivada fxx e em D
print('Testando o ponto (-1,-1):')
print('fxx =',z.diff(x,x).subs([(x, -1), (y, -1)]))
print('D =',D.subs([(x, -1), (y, -1)]))
```

```python
#substituindo x=1 e y = 1 na derivada fxx e em D
print('Testando o ponto (1,1):')
print('fxx =',z.diff(x,x).subs([(x, 1), (y, 1)]))
print('D =',D.subs([(x, 1), (y, 1)]))
    
```
