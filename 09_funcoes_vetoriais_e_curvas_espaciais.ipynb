## 9. Funções Vetoriais e Curvas Espaciais

Uma função vetorial, ou função com valores vetoriais, é uma função cujo domínio é um conjunto de números reais e cuja imagem é um conjunto de vetores ([Stewart, 2022](https://integrada.minhabiblioteca.com.br/reader/books/9786555584103/pageid/301)).

Para funções vetoriais $\mathbf{r}$ cujos valores são vetores tridimensionais, para todo número $t$ no domínio de $\mathbf{r}$ existe um único vetor de $V_3$ denotado por $\mathbf{r}(t)$. Se $f(t), g(t)$ e $h(t)$ são as componentes do vetor $\mathbf{r}(t)$, então $f, g$ e $h$ são funções com valores reais chamadas funções. Denotamos por

$$
\mathbf{r}(t)=\langle f(t), g(t), h(t)\rangle=f(t) \mathbf{i}+g(t) \mathbf{j}+h(t) \mathbf{k}
$$

As curvas espaciais e as funções vetoriais contínuas estão intimamente relacionadas. Suponha que $f, g$ e $h$ sejam funções reais contínuas em um intervalo $I$. Em seguida, o conjunto $C$ de todos os pontos $(x, y, z)$ no espaço, sendo

$$
x=f(t) \quad y=g(t) \quad z=h(t)
$$

e $t$ varia no intervalo $I$, é chamado **curva espacial**. As equações acima são denominadas equações paramétricas de $C$ e $t$ é conhecido como parâmetro.

**Exemplo 9.3:** ([Stewart, 2022, p. 824](https://integrada.minhabiblioteca.com.br/reader/books/9786555584103/pageid/303)) Determine uma equação vetorial que represente a curva obtida pela intersecção do cilindro $x^2+y^2=1$ com o plano $y+z=2$.

Podemos escrever as equações paramétricas para $C$ como

$$
x=\cos t \quad y=\operatorname{sen} t \quad z=2-\operatorname{sen} t \quad 0 \leq t \leq 2 \pi
$$

então, a equação vetorial correspondente é 

$$\mathbf{r}(t)=\cos t \mathbf{i}+\operatorname{sen} t \mathbf{j}+(2-\operatorname{sen} t) \mathbf{k} \quad 0 \leq t \leq 2 \pi$$

```python
#cria a figura
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')

# cilindro - coordenadas cilindricas
t,z = np.meshgrid(np.linspace(0,2*np.pi,36), np.linspace(0,5,20))
x = np.cos(t)
y = np.sin(t)
ax.plot_wireframe(x, y, z, lw=0.2, color='k')

# plano
x,y = np.meshgrid(np.linspace(-1,1,20), np.linspace(-1,1,20))
z = 2-y
ax.plot_wireframe(x, y, z, lw=0.2, color='k')

t = np.linspace(0,2*np.pi,36)
x = np.cos(t)
y = np.sin(t)
z = 2-np.sin(t)
ax.plot(x, y, z, lw=1, color='r')
```

**Exemplo 9.4:** ([Stewart, 2022, p. 825](https://integrada.minhabiblioteca.com.br/reader/books/9786555584103/pageid/303)) Determine as equações paramétricas da curva formada pela intersecção do paraboloide $4 y=x^2+z^2$ com o plano $y=x$.

*SOLUÇÃO:* Como qualquer ponto da curva $C$, que está na intersecção das superfícies, satisfaz as equações das duas superfícies, podemos substituir $y=x$ na equação do paraboloide, o que nos fornece $4 x=x^2+z^2$. Completando o quadrado em $x$, obtemos $(x-2)^2+$ $z^2=4$, de modo que $C$ deve estar contida no cilindro circular $(x-2)^2+z^2=4$ e a projeção de $C$ no plano $x z$ será a circunferência definida por $(x-2)^2+z^2=4, y=0$, que tem centro em $(2,0,0)$ e raio 2. Assim, podemos escrever $x=2+2 \cos t$, $z=2 \operatorname{sen} t, 0 \leq t \leq 2 \pi$, e, uma vez que $y=x$, as equações paramétricas de $C$ são

$$
x=2+2 \cos t, \quad y=2+2 \cos t, \quad z=2 \operatorname{sen} t, \quad 0 \leq t \leq 2 \pi
$$


```python
#%matplotlib tk
%matplotlib inline
```

```python
#cria a figura
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')

# paraboloide
t = np.linspace(0, 2*np.pi, 37)
r = np.linspace(0, 5)
rv, tv = np.meshgrid(r, t)

x = rv*np.cos(tv)
z = rv*np.sin(tv)
y = x**2/4 + z**2/4

#ax.plot_surface(x, y, z, alpha=0.5)
ax.plot_wireframe(x, y, z, lw=0.2, color='k')

# plano
x = np.linspace(0,5,21)
y = x.copy()
z = np.repeat(np.array([np.linspace(-5,5,11)]), len(x), axis=0).T

#ax.plot_surface(x, y, z, alpha=0.5)
ax.plot_wireframe(x, y, z, lw=0.2, color='k')

# curva
t = np.linspace(0, 2*np.pi, 37)

x = 2+2*np.cos(t)
y = 2+2*np.cos(t)
z = 2*np.sin(t)

ax.plot(x, y, z, lw=1, color='r')


ax.set_box_aspect([1,1,1])
plt.show()
```

### 9.2 Plotando campos vetoriais 

([Stewart, 2016](https://integrada.minhabiblioteca.com.br/reader/books/9786555584103/pageid/518)) Seja $D$ um conjunto em $\mathbb{R}^2$ (uma região plana). Um campo vetorial em $\mathbb{R}^2$ é uma função $\mathbf{F}$ que associa a cada ponto $(x, y)$ em $D$ um vetor bidimensional $\mathbf{F}(x, y)$.

Uma vez que $\mathbf{F}(x, y)$ é um vetor bidimensional, podemos escrevê-lo em termos de suas funções componentes $P$ e $Q$ da seguinte forma:

$$
\mathbf{F}(x, y)=P(x, y) \mathbf{i}+Q(x, y) \mathbf{j}=\langle P(x, y), Q(x, y)\rangle
$$

ou, de forma mais compacta,

$$
\mathbf{F}=P \mathbf{i}+Q \mathbf{j}
$$


Observe que $P$ e $Q$ são funções escalares de duas variáveis e são chamadas, algumas vezes, campos escalares, para distingui-los dos campos vetoriais.

**Exemplo 9.7:**
Nesse exemplo vamos representar graficamente o campo de vetores dados pela função $$f(x,y)=\left(\frac{-y^2}{x^4+y^4},\frac{x}{\sqrt{x^2+y^2}} \right)$$

usando uma malha de 10x10 pontos, com $-5 \leq x \leq 5$ e $-5 \leq y \leq 5$.

```python
fig = plt.figure(figsize=(4,3))
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)
xi,yi = np.meshgrid(x,y)

u = -yi**2/np.sqrt(xi**4 + yi**4)
v = xi/np.sqrt(xi**2 + yi**2)

#plt.streamplot(xi,yi,u,v)

plt.quiver(xi,yi,u,v)
plt.show()
```

Seja $E$ um subconjunto de $\mathbb{R}^3$. Um campo vetorial em $\mathbb{R}^3$ é uma função $\mathbf{F}$ que associa a cada ponto $(x, y, z)$ em $E$ um vetor tridimensional $\mathbf{F}(x, y, z)$.

Podemos escrever um campo vetorial $\mathbf{F}$ em $\mathbb{R}^3$ em termos das funções componentes $P, Q$ e $R$ como

$$
\mathbf{F}(x, y, z)=P(x, y, z) \mathbf{i}+Q(x, y, z) \mathbf{j}+R(x, y, z) \mathbf{k}
$$


**Exemplo 9.8:** Nesse exemplo vamos representar graficamente o campo de vetores dados pela função $$f(x,y,z)=\left( sen(x)cos(yz),-yz, cos(x)sen(z) \right)$$

usando uma malha de 10x10 pontos, com $-5 \leq x \leq 5$ e $-5 \leq y \leq 5$.

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection='3d')

x,y,z = np.meshgrid(np.arange(-1, 1, 0.25),
                    np.arange(-1, 1, 0.25),
                    np.arange(-1, 1, 0.25))

u = np.sin(x) * np.cos(y*z)
v = -y*z
w = np.cos(x) * np.sin(z)

ax.quiver(x, y, z, u, v, w, 
          length=0.2, linewidths= 0.7,
          color = 'black')

plt.show()
```

### 9.3 Campo gradiente
Se $f$ é uma função escalar de duas variáveis, seu gradiente $\nabla f$ (ou $\operatorname{grad} f$ ) é definido por

$$
\nabla f(x, y)=f_x(x, y) \mathbf{i}+f_y(x, y) \mathbf{j}
$$

Portanto, $\nabla f$ é realmente um campo vetorial em $\mathbb{R}^2$ e é denominado campo vetorial gradiente. Da mesma forma, se $f$ for uma função escalar de três variáveis, seu gradiente é um campo vetorial em $\mathbb{R}^3$ dado por

$$
\nabla f(x, y, z)=f x(x, y, z) \mathbf{i}+f y(x, y, z) \mathbf{j}+f z(x, y, z) \mathbf{k}
$$



**Exemplo 9.9:** Exemplo de gráficos de campos gradiente (campo vetorial) associado a uma função de duas variáveis (campo escalar).

```python
f1 = lambda x,y: x**2+y**2
f2 = lambda x,y: x**2-y**2
f3 = lambda x,y: x**4+y**4-4*x*y+1
```

```python
xi, yi = np.meshgrid(np.linspace(-5,5,201),
                     np.linspace(-5,5,201))

#criamos a figura e o subplot 3d
fig = plt.figure(figsize=(13,4))

ax = fig.add_subplot(131, projection='3d')
ax.plot_surface(xi, yi, f1(xi,yi))

ax = fig.add_subplot(132, projection='3d')
ax.plot_surface(xi, yi, f2(xi,yi))

ax = fig.add_subplot(133, projection='3d')
ax.plot_surface(xi, yi, f3(xi,yi))

plt.show()
```

```python
xi, yi = np.meshgrid(np.linspace(-5,5,21),
                     np.linspace(-5,5,21))

#criamos a figura e o subplot 3d
fig = plt.figure(figsize=(13,4))

ax = fig.add_subplot(131)
u = np.gradient(f1(xi,yi))[1]
v = np.gradient(f1(xi,yi))[0]
plt.quiver(xi,yi,u,v, color='b')
plt.contour(xi, yi, f1(xi,yi),10, colors="r", linewidths=0.4)


ax = fig.add_subplot(132)
u = np.gradient(f2(xi,yi))[1]
v = np.gradient(f2(xi,yi))[0]
plt.quiver(xi,yi,u,v, color='b')
plt.contour(xi, yi, f2(xi,yi),10, colors="r", linewidths=0.4)


ax = fig.add_subplot(133)
u = np.gradient(f3(xi,yi))[1]
v = np.gradient(f3(xi,yi))[0]
plt.quiver(xi,yi,u,v, color='b')
xi, yi = np.meshgrid(np.linspace(-5,5), np.linspace(-5,5))
niveis = [-0.9, 0, 1, 2,10,50,100, 200, 400, 800]
plt.contour(xi, yi, f3(xi,yi),niveis, colors="r", linewidths=0.4)

plt.show()
```
